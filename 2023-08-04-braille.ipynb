```python
from dataclasses import dataclass
```

```python
@dataclass
class Braille:
    x: int
    rows: int
    cols: int
        
    def get_index(self, r, c):
        return self.cols * r + c
    
    def pretty(self):
        out = ""
        for r in range(self.rows):
            for c in range(self.cols):
                if self.x & (1 << self.get_index(r, c)):
                    out += "#"
                else:
                    out += "."
            out += "\n"
        return out
        
    @classmethod
    def generate_all(cls, rows=3, cols=2):
        nbits = rows * cols
        for i in range(0, 2 ** nbits):
            yield cls(i, rows, cols)
            
for c in Braille.generate_all():
    print(c)
    print(c.pretty())
```
